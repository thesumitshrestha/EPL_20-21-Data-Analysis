```python
# Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Reading dataset
df = pd.read_csv('EPL_20_21 2.csv')
```

```python
#Checking null values
df.isna().sum()
```

```python
# Checking the clubs
df['Club'].unique()
```

```python
# Creating 2 new columns

df['MinsPerMatch'] = (df['Mins']/df['Matches']).astype(int)
df['GoalsPerMatch'] = (df['Goals']/df['Matches']).astype(float)
```

```python
df.head()
```

```python
# Red card from Manchester United
df[(df['Club'] == 'Manchester United') & (df['Red_Cards'] > 0)]
```

```python
df.groupby(['Goals', 'Name']).sum()
```

```python
# Total Goals

total_goals = df['Goals'].sum()
total_goals
```

```python
# Penalty Goals

total_penalty_goals = df['Penalty_Goals'].sum()
total_penalty_goals
```

```python
# Penalty Attempts

total_penalty_attempts = df['Penalty_Attempted'].sum()
total_penalty_attempts
```

```python
# Piechart for penalty missed vs scored

plt.figure(figsize=(13, 6))
pl_not_scored = total_penalty_attempts - total_penalty_goals
data = [pl_not_scored, total_penalty_goals]
labels = ['Penalties Missed', 'Penalty Scored']
color = sns.color_palette('Set2')

plt.pie(data, labels=labels, colors=color, autopct='%.0f%%')
plt.title('Penalties Missed vs Penalties Scored')
plt.axis('equal')
plt.show()
```

```python
# Unique positions

df['Position'].unique()
```

```python
# Total FW Players
df[df['Position'] == 'FW']
```

```python
# Players from different nations
np.size(df['Nationality'].unique())
```

```python
# Most player from which countries
top_nationality = df.groupby('Nationality').size().sort_values(ascending=False)[:1]
print(top_nationality)

nationality = df.groupby('Nationality').size().sort_values(ascending=False)
nationality.head(10).plot(kind='bar', figsize=(12,6), color=sns.color_palette('magma')).grid()
```

```python
# Clubs with maximum number of players in the squad

df['Club'].value_counts().nlargest(5).plot(kind='bar', color=sns.color_palette('Set2'))
```

```python
# Clubs with least players in the squad

df['Club'].value_counts(ascending=True).nsmallest(5).plot(kind='bar', color=sns.color_palette('magma')).grid()
```

```python
# Players based on age group
under20 = df[df['Age'] <= 20]
age20_25 = df[(df['Age'] > 20) & (df['Age'] <=25)]
age25_30 = df[(df['Age'] > 25) & (df['Age'] <=30)]
above30 = df[df['Age']>30]
```

```python
x = np.array([under20['Name'].count(), age20_25['Name'].count(), age25_30['Name'].count(), above30['Name'].count()])
myLabels = ["<=20", ">20 & <=25", ">25 & <=30", ">30"]
plt.title("Players based on age group")
plt.pie(x, labels=myLabels, autopct="%.1f%%")
plt.show()
```

```python
# Total under 20 playes in the club

player_u_20 = df[df['Age'] < 20]
player_u_20['Club'].value_counts().plot(kind='bar')
```

```python
# Under 20 Players in Manchester United

player_u_20[player_u_20['Club'] == 'Manchester United']
```

```python
# Under 20 players in Chelsea

df[(df['Age'] < 20) & (df['Club'] == 'Chelsea')]
```

```python
# Average age of players in each club

plt.figure(figsize=(12,6))
sns.boxplot(x ='Club', y='Age', data = df)
plt.xticks(rotation=90)
plt.show()
```

```python
num_player = df.groupby('Club').size()
avg_player = (df.groupby('Club')['Age'].sum())/ num_player
avg_player.sort_values(ascending=False)
```

```python
# Total assists from each club

total_assists_club = df.groupby('Club')['Assists'].sum()
total_assists_club.plot(kind='bar', color=sns.color_palette('Set1'))
```

```python
# Top 10 assists player list

top_players_assist = df[['Name', 'Club', 'Assists', 'Matches']].nlargest(10, columns='Assists')
top_players_assist
```

```python
# Top 10 highest scorer

highest_scorer = df[['Name', 'Club', 'Goals', 'Matches']].nlargest(10, columns='Goals')
highest_scorer
```

```python
# Goals per match
goals_per_match = df[['Name', 'GoalsPerMatch', 'Goals', 'Matches']].nlargest(10, columns='GoalsPerMatch')
goals_per_match
```

```python
# Top 10 players with most yellow card

top_10_yellow_cards = df[['Name', 'Club', 'Yellow_Cards']].nlargest(10, columns='Yellow_Cards')
top_10_yellow_cards
```
